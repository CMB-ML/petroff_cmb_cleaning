```python
from pathlib import Path
import numpy as np
import matplotlib.pyplot as plt
import healpy as hp
import colorcet as cc
from healpy import projaxes as PA
```

```python
# Modify Healpy's `mollview` function to allow more control of margins
def mollview(
    bounds,
    map=None,
    unit="",
    xsize=800,
    title="",
    min=None,
    max=None,
    cbar=True,
    cmap=None,
    extend='neither',
):
    # Ensure that the nside is valid
    nside = hp.get_nside(map)

    f = plt.gcf()
    left, bottom, right, top = bounds
    extent = (left, bottom, right - left, top - bottom)

    map = hp.pixelfunc.ma_to_array(map)
    ax = PA.HpxMollweideAxes(f, extent)
    f.add_axes(ax)
    img = ax.projmap(map, xsize=xsize, vmin=min, vmax=max, cmap=cmap, badcolor="lightgray")
    if cbar:
        im = ax.get_images()[0]
        b = im.norm.inverse(np.linspace(0, 1, im.cmap.N + 1))
        v = np.linspace(im.norm.vmin, im.norm.vmax, im.cmap.N)
        cb = f.colorbar(
            im,
            ax=ax,
            orientation="horizontal",
            shrink=0.5,
            aspect=25,
            ticks=PA.BoundaryLocator(2, None),
            pad=0.05,
            fraction=0.1,
            boundaries=b,
            values=v,
            extend=extend
        )
        cb.solids.set_rasterized(True)
        cb.ax.text(0.5, -1.0, unit, transform=cb.ax.transAxes, ha="center", va="top")
    ax.set_title(title)
    f.sca(ax)
```

```python
plt.rc('text', usetex=True)
plt.rc('text.latex', preamble=r'\usepackage{siunitx}\sisetup{detect-all}\usepackage{sansmath}\sansmath')
params = {
    'legend.fontsize': 9,
    'axes.labelsize': 9,
    'xtick.labelsize': 9,
    'ytick.labelsize': 9,
}
plt.rcParams.update(params)
```

```python
# Data locations
psm_dir = Path('../simulations/outputnops512_dm8')
commander_map_path = '/data/planck/COM_CMB_IQU-commander_2048_R3.00_full.fits'
```

```python
# Beam window (arcmin)
beam_window = 13.1
```

## Simulation map from test set

```python
y_raw = hp.read_map(psm_dir / f'sim0999/cmb.fits', verbose=False)
```

```python
# Load results
result = hp.read_map('sim-ann-map.fits', verbose=False)
result_err = hp.read_map('sim-ann-map-err.fits', verbose=False)
```

```python
plt.figure(figsize=(9.0, 6.0))

f = plt.gcf()
f.delaxes(f.gca())

mollview((0.01, 0.5, 0.49, 1.0), y_raw, min=-352.8, max=352.8, unit='\si{\micro\kelvin}', title='Reference', cmap='cet_bjy', extend='both')

# Draw border: this could be done using hp.graticule, but it has slight
# rendering artifacts at top and bottom of border
theta = np.linspace(0, np.pi, 300)
plt.gca().projplot(np.append(theta[::-1], theta), np.append(np.repeat(-np.pi, 300), np.repeat(0.999999 * np.pi, 300)), "-k", lw=1, direct=True, solid_capstyle='round', solid_joinstyle='round')

mollview((0.01, 0.0, 0.49, 0.5), result, min=-352.8, max=352.8, unit='\si{\micro\kelvin}', title='Reconstruction', cmap='cet_bjy', extend='both')
plt.gca().projplot(np.append(theta[::-1], theta), np.append(np.repeat(-np.pi, 300), np.repeat(0.999999 * np.pi, 300)), "-k", lw=1, direct=True, solid_capstyle='round', solid_joinstyle='round')

mollview((0.51, 0.5, 0.99, 1.0), result - y_raw, min=-50.4, max=50.4, unit='\si{\micro\kelvin}', title='Difference', cmap='cet_bjy', extend='both')
plt.gca().projplot(np.append(theta[::-1], theta), np.append(np.repeat(-np.pi, 300), np.repeat(0.999999 * np.pi, 300)), "-k", lw=1, direct=True, solid_capstyle='round', solid_joinstyle='round')

mollview((0.51, 0.0, 0.99, 0.5), result_err, min=0, max=50.2, unit='\si{\micro\kelvin}', title='Predicted Error', cmap='cividis', extend='max')
plt.gca().projplot(np.append(theta[::-1], theta), np.append(np.repeat(-np.pi, 300), np.repeat(0.999999 * np.pi, 300)), "-k", lw=1, direct=True, solid_capstyle='round', solid_joinstyle='round')

plt.savefig('sim-maps.pdf', dpi=300)
```

## Evaluation using Planck maps

```python
# Load results
result_planck = hp.read_map('planck-ann-map.fits', verbose=False)
result_planck_err = hp.read_map('planck-ann-map-err.fits', verbose=False)
```

```python
strip_mask = np.zeros(hp.nside2npix(512), dtype=np.bool)
strip_mask[hp.query_strip(512, np.deg2rad(100), np.deg2rad(80))] = True
strip_mask = hp.reorder(strip_mask, r2n=True)
result_median = np.nanmedian(result_planck[strip_mask])
print(result_median)
```

```python
commander_map = hp.ud_grade(hp.smoothing(hp.read_map(commander_map_path, field=0), np.deg2rad(beam_window / 60), use_pixel_weights=True), 512) * 1e6
```

```python
plt.figure(figsize=(9.0, 6.0))

f = plt.gcf()
f.delaxes(f.gca())

mollview((0.01, 0.5, 0.49, 1.0), commander_map, min=-352.8, max=352.8, unit='\si{\micro\kelvin}', title='Commander', cmap='cet_bjy', extend='both')

# Draw border: this could be done using hp.graticule, but it has slight
# rendering artifacts at top and bottom of border
theta = np.linspace(0, np.pi, 300)
plt.gca().projplot(np.append(theta[::-1], theta), np.append(np.repeat(-np.pi, 300), np.repeat(0.999999 * np.pi, 300)), "-k", lw=1, direct=True, solid_capstyle='round', solid_joinstyle='round')

mollview((0.01, 0.0, 0.49, 0.5), result_planck - result_median, min=-352.8, max=352.8, unit='\si{\micro\kelvin}', title='Reconstruction', cmap='cet_bjy', extend='both')
plt.gca().projplot(np.append(theta[::-1], theta), np.append(np.repeat(-np.pi, 300), np.repeat(0.999999 * np.pi, 300)), "-k", lw=1, direct=True, solid_capstyle='round', solid_joinstyle='round')

mollview((0.51, 0.5, 0.99, 1.0), result_planck - result_median - commander_map, min=-50.4, max=50.4, unit='\si{\micro\kelvin}', title='Difference', cmap='cet_bjy', extend='both')
plt.gca().projplot(np.append(theta[::-1], theta), np.append(np.repeat(-np.pi, 300), np.repeat(0.999999 * np.pi, 300)), "-k", lw=1, direct=True, solid_capstyle='round', solid_joinstyle='round')

mollview((0.51, 0.0, 0.99, 0.5), result_planck_err, min=0, max=50.2, unit='\si{\micro\kelvin}', title='Predicted Error', cmap='cividis', extend='max')
plt.gca().projplot(np.append(theta[::-1], theta), np.append(np.repeat(-np.pi, 300), np.repeat(0.999999 * np.pi, 300)), "-k", lw=1, direct=True, solid_capstyle='round', solid_joinstyle='round')

plt.savefig('planck-maps.pdf', dpi=300)
```
